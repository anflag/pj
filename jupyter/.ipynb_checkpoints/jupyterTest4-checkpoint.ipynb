```python
import copy
pre_data02 = copy.deepcopy(pre_data)
data03 = pd.read_pickle("../datasets/electric/201803_pay_from_next_month_data.pkl")
pre_data03 = data03[['user_id','fee','payment_time_2','pay_days']]
pre_data03.columns = ['user_id','fee03','pay_time03','pay_days03']
pre_data03.head()
```

```python
temp = pre_data02.merge(pre_data03, on='user_id', how='outer')
temp
```
