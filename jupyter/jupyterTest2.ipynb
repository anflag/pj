```python
import pandas as pd
data = pd.read_pickle("../datasets/electric/201803_data.pkl")
data.head()
```

```python
data = pd.read_pickle("../datasets/electric/201803_notna_date_data.pkl")
data.info()
```

```python
# 4月1日之后缴费的
pay_from_next_month_data = data[data['pay_days']>27]
pay_from_next_month_data
```

```python
pay_from_next_month_data.info()
```

```python
pay_days_data = pay_from_next_month_data['pay_days'].astype(int).values
pay_days_data.max()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(20, 8), dpi=80)
num_bin = (pay_days_data.max() - pay_days_data.min())
plt.hist(pay_days_data,num_bin)
```
